```python
import numpy as np
import pandas as pd
import requests
```

```python
data = pd.DataFrame()
for i in range(1,472):
  response = requests.get(f'https://api.themoviedb.org/3/movie/top_rated?api_key=8265bd1679663a7ea12ac168da84d2e8&language=en-US&page={i}')
  temp = pd.DataFrame(response.json()['results'])[['id','genre_ids','title','overview','release_date','popularity','vote_average','vote_count']]
  data = pd.concat([data, temp], ignore_index=True)
```

```python
genres = requests.get('https://api.themoviedb.org/3/genre/movie/list?api_key=8265bd1679663a7ea12ac168da84d2e8&language=en-US')
print(genres.json())
```

```python
g = genres.json()['genres']
```

```python
g
```

```python
genre_dict = {genre['id']: genre['name'] for genre in g}
print(genre_dict)
```

```python
def replace_genre_ids_with_names(genre_ids):
    return [genre_dict[genre_id] for genre_id in genre_ids]
```

```python
# Apply the function to the genre_ids column
data['genre_names'] = data['genre_ids'].apply(replace_genre_ids_with_names)
```

```python
data.head()
```

```python
data = data.drop(columns=['genre_ids', 'release_date', 'popularity', 'vote_average', 'vote_count'], axis=1)
```

```python
data['overview'] = data['overview'].str.lower()
```

```python
data.head()
```

```python
import string
exclude = string.punctuation
def remove_punct(text):
  return text.translate(str.maketrans('','',exclude))
```

```python
data['overview'] = data['overview'].apply(remove_punct)
```

```python
data.head()
```

```python
import nltk
nltk.download('stopwords')
```

```python
from nltk.corpus import stopwords
```

```python
stopwords.words('english')
```

```python
def remove_stop(text):
  new_text = []
  for word in text.split():
    if word in stopwords.words('english'):
      new_text.append('')
    else:
      new_text.append(word)
  return " ".join(new_text)
```

```python
data['overview'] = data['overview'].apply(remove_stop)
```

```python
import nltk
from nltk.stem import PorterStemmer
```

```python
ps = PorterStemmer()
def stemming(text):
  return " ".join([ps.stem(word) for word in text])
```

```python
data['overview'] = data['overview'].apply(lambda x: stemming(x))
```
